### 建立資料庫
- 1.創建資料庫
- 2.建立資料表
- 3.寫入不重複資料
- 4.取最新數據

```python
import sqlite3
```

```python
conn=sqlite3.connect('pm25.db')
cursor=conn.cursor()

conn,cursor
```

```python
datas[0].keys()
```

```python
sqlstr='''
create table if not exists data(
id integer primary key autoincrement,
site text,
county text,
pm25 integer,
datacreationdate datetime,
itemunit text
)


'''
cursor.execute(sqlstr)
conn.commit()
```

```python
values = []
for data in datas:
    data = list(data.values())
    try:
        data[2] = eval(data[2])
        values.append(data)
    except Exception as e:
        print(e)
values
```
